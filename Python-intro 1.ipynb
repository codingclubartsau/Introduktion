Før at du kan komme i gang, skal du have en tekst at arbejde med. Vi træner på en tekst med Tessas nytårstale fra www.dansketaler.dk. De følgende punkter fortæller fremgangsmåden, hvis man gerne vil hente tekstfiler ind i sit program - her henter vi altså Tessas nytårstale ind i programmet, så vi kan arbejde med den:

* Gå ind på https://dansketaler.dk/tale/tessas-nytaarstale-2021/
* Kopier alt teksten på siden
* Åbn programmet Notesblok på Windows eller TextEdit på Mac
* Sæt den kopierede tale ind i noteprogrammet. (Det skal gerne være ren tekst, uden en masse symboler/forstyrrende elementer)
* Gem din tekstfil under navnet "tessa" i den mappe samme mappe som denne jupyter-notebook-fil ligger gemt (dette filnavn slutter med .ipynb). Når du gemmer din tekstfil, skal den gerne have endelsen ".txt" (wordfiler har typist .docx, pdf-filer har .pdf osv. Det fortæller hvilken filtype din fil er, og ".txt" betyder, at du har gemt din fil som ren tekst uden nogen formatering)

```python
txt = open("tessa.txt", encoding="utf8") #etablerer en sti hen til filens placering
tale = txt.read()
```

OBS: Hvis du prøver ovenstående, og det ikke virker, kan du evt. kopiere teksten fra talen direkte ind under et variabelnavn, og slette alle linjeskift. Eksempelvis som:
tale = "indsæt tekst her og fjern linjeskift"
Dette kan du gøre i kodefeltet nedenfor

Nu har vi hentet talen og gemt den under variabelnavnet "tale". Prøv at bruge funktionen "print" nedenfor, og se hvad der sker.

```python
print(tale)
```

Hvis vi bare skriver "tale" i kodefeltet, kommer det, der er gemt under variablen "tale" også frem, men der er linjeskift gemt under formateringskoden "\n", og det kan se rodet ud, når man læser den. Prøv selv at se:

```python
tale
```

Fordi det ser rodet ud at vi får koden med for formatering, plejer vi altså at bruge "print", når vi gerne vil se teksten som vi arbejder med.

Vi vil gerne prøve at dele tekten op i sætninger. Dette kan vi gøre ved at dele teksten op i strings (strenge af tekst, som er omkranset med citationstegn " "). Dette kan vi gøre ved at bruge ".split(".")"-funktionen på teksten. Syntaksen fungerer sådan, at man først skriver navnet på den tekst man vil bruge funktionen på, og bagefter skriver man ".split(".")".

```python
tale_sætninger = tale.split(".")
print(tale_sætninger)
```

Det, der står i parentesen ved ".split()", er det som vi deler teksten op i. Prøv eksempelvis at bytte ud med "," eller "o" eller dit yndlingstegn, og se hvad der sker.

Vi kan nu se hvor mange sætninger der er i alt i talen. Det kan vi gøre med "len()". "len" er kort for 'length', og inde i parentesen skal du skrive det variabelnavn, som du gerne vil måle længden af.

```python
len(tale_sætninger)
```

Vi ved altså hvor mange sætninger der er. Når det er en så kort tekst som Tessas tale, kræver det ikke så meget arbejde at sided og tælle sætningerne manuelt, men tænk hvis vi ville tælle alle sætningerne i en roman, eller alle sætninger i et helt forfatterskab? Så er det ret smart at kunne bruge "len()", i stedet for at skulle tælle det hele manuelt!

Vi vil nu gerne prøve at tælle hvor lang hver sætning er i talen Det kan vi også bruge "len()" til, men nu skal vi bruge den på hver enkelt sætning i stedet for på listen med sætninger. Til det kan vi bruge det der hedder et "for-loop". Her er syntaksen at vi siger "for alle sætninger i listen af sætninger: tæl længden på sætningen"", og det ser ud således, når man skriver det i kode:

```python
for sætning in tale_sætninger:
    print(len(sætning))
```

Nu kan vi se hvor mange antal tegn der er i hver sætning. Hvis vi tæller efter, kan vi dog se at nogle sætninger lige nu har for mange tegn ift. det som vi ser i talen. Det er fordi vi stadig har "\n"-formateringen gemt i vores tale, hver gang at der er et linjeskift i talen. (Det kan du se, ved at køre næste linje kode, hvis du lige har glemt hvordan det så ud).

```python
tale_sætninger
```

Vi vil altså gerne have "renset" vores tekst for linjeskift og overflødige tegn, da vi ikke er interesserede i at kigge på tegnene i den her undersøgelse af teksten. Det kan vi gøre ved funktionen ".replace()", hvor man kan erstatte et tegn med et andet. Vi kan altså eksempelvis erstatte punktummer med mellemrum, og så vil vi skrive funktionen som ".replace("."," ")".
Prøv at køre nedenstående kodefelt, og se hvad der sker med teksten, når du printer den.

```python
tale_clean = tale.replace("\n", " ").replace("-", " ").replace(":", " ").replace("’"," ").replace(".", " ").replace(",", " ").replace("”", " ").replace(";", " ").replace("“", " ").replace("–", " ").replace("…", " ")
tale_clean
```

Vi kan nu lave en ny sætningsliste, hvor hver sætning i talen omkranses af tegnene "[ ]". Det kalder man en liste i python.
Først laver vi en tom liste, der hedder sætningsliste. Herefter laver vi igen et "for-loop": "for alle sætninger i 'tale_sætninger' skal du gøre følgende (og nu læser vi bagfra i koden!)
* split sætningen op ved mellemrum (fordi der ikke står noget i parentesen ved ".split()", er det mellemrum pr. default)
* tilføj nu delene af sætningen til vores sætningsliste (med sætningsliste.append())

```python
sætningsliste = []
for sætning in tale_sætninger:
    sætningsliste.append(sætning.split())
    
sætningsliste
```

Nu har vi ingen forstyrrende tegn i sætningslisten!

Vi kan nu tælle hvor mange ord der er i hver sætning i talen. Det gør vi ved at bruge "len()" igen.

```python
for sætning in sætningsliste:
    print(len(sætning))
```

Vi kan også se hvor mange ord der er i hele talen. Det kan vi gøre ved at dele teksten op ved alle mellemrum i stedet for at dele teksten op ved punktummer. Vi bruger altså funktionen ".split()" igen. Den funktion virker automatisk (pr. default) for mellemrum, hvis man ikke beder den eksplicit om at gøre noget andet (f.eks. punktum i stedet).

```python
ordliste = tale_clean.split()
print(ordliste)
```

Nu kunne det være spændende at se, hvilke ord der bruges mest i talen. Til det kan vi hente en smart tælle-funktion fra det modul der hedder "Counter". For at kunne bruge funktionen skal vi først hente den ind i vores program. Det kan vi gøre ved at skrive følgende kode:

```python
from collections import Counter
```

Nu kan vi bruge funktionen Counter på vores ordliste. Vi gemmer værdierne fra funktionen under variablen "c":

```python
c = Counter(ordliste)
c
```

Nu kan vi se hvor mange gange hvert ord optræder i teksten; 'Godaften' forekommer 1 gang, 'Jeg' forekommer 7 gange osv.
